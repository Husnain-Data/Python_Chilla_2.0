Data Cleaning using Pandas.

Data cleaning means fixing bad data in your data set.


Bad data could be:

1-Empty cells
2- Data in wrong format
3- Wrong data
4- Duplicates

```python
d_data = pd.read_csv('dirtydata.csv')
d_data.head(5)
```

```python
#no empty cells
n_data = d_data.dropna()
n_data.head()
```

```python
#removing null values
n_data = d_data.dropna( inplace=True)
n_data
```

```python
d_data.info()
```

```python
d1_data = pd.read_csv("dirtydata.csv")
d1_data.head(10)
```

```python
#Filling null values
d1_data["Calories"].fillna(d1_data["Calories"].mean(), inplace= True)
```

```python
#Correcting Format
pd.to_datetime(d_data["Date"])
d1_data.info()
```

```python
#Removing Rows
d_data.dropna(subset=['Date'], inplace=True)
d_data.info()
```

```python
#Using specific row to replace data

d1_data.loc[7, "Duration"] = 45
d1_data.loc[7]
```

```python
#Using loop to restrict data

for x in d1_data.index:
    if d1_data.loc[x, "Duration"] > 120:
        d1_data.loc[x, "Duration"] = 120

d1_data["Duration"].min()
```

```python
#Dropping rows contain wrong data

for x in d1_data.index:
    if d1_data.loc[x, "Duration"] > 120:
        d1_data.drop(x, inplace = True)
d1_data.info()
```

```python
#Checking duplication
s = d1_data.duplicated().sum()
s
```

```python
a = d1_data.drop_duplicates()
s = a.duplicated().sum()
a1 = a.dropna()
a1.to_csv("cleandata1.csv")
```
