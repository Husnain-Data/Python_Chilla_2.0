```python
#16 - Spllting String into multiple columns
df = pd.DataFrame({'name':['Muhammad Husnain', 'Muhammad Mushhood', 'Zahoor Ahmad'], 
                   'location':['Lahore Pakistan', 'Toba Tek Singh', 'Punjab Pakistan']})
df
```

```python
#Using Expand method

df.location.str.split(' ', expand=True).head()
```

```python
#updating column

df[['city', 'country', 'suffix']] = df.location.str.split(' ', expand=True)
df
```

```python
df[['First Name', 'Last Name']] = df.name.str.split(' ', expand=True)
df
```

```python
#refine data (Manipulation of Data)
df[["First Name", "Last Name", "City", "Country/City Name", "Suffix"]] = df[['First Name', 'Last Name', 'city', 'country', 'suffix']]
df = df[["First Name", "Last Name", "City", "Country/City Name", "Suffix"]]
df
```
