```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep="|")
users
```

### Step 4. Discover what is the mean age per occupation

```python
users['age'].groupby(users['occupation']).mean()
age_mean = pd.DataFrame({'age_mean': users['age'].groupby(users['occupation']).mean()})

age_mean
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
Male_ratio= pd.DataFrame({'Male Mean per Occupation':users['gender_M'].groupby(users['occupation']).mean().sort_values(ascending=False)})
Male_ratio
```
